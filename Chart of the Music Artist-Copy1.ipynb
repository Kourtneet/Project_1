```python
all_artist_df = pd.concat([Mariahdata_df, Taydata_df, beydata_df],ignore_index=True)
all_artist_df
```

```python
all_artist_df_sorted = all_artist_df.sort_values(by=['popularity', 'speechiness', 'loudness', 'danceability'],ascending=False, ignore_index=True)
all_artist_df_sorted
```

```python
# Build the scatter plots for all_artist_df["beydata_df"], all_artist_df["popularity"], marker="o"
plt.scatter(all_artist_df_sorted['popularity'], all_artist_df_sorted['tempo'], marker="o")

# Incorporate the other graph properties
plt.title("Tempo vs Popularity for All Artists")
plt.ylabel("tempo")
plt.xlabel("popularity")
plt.grid(True)
#colors = np.where(x > 0.5, 'green', 'red', 'blue')

correlation = st.pearsonr(all_artist_df_sorted['tempo'], all_artist_df_sorted['popularity'])
print(f"The correlation between Tempo and Popularity is {round(correlation[0],2)}")

pe_slope, pe_int, pe_r, pe_p, pe_std_err = st.linregress(all_artist_df_sorted['popularity'], all_artist_df_sorted['tempo'])
pe_fit = pe_slope * all_artist_df_sorted['popularity']+ pe_int
plt.plot(all_artist_df_sorted['popularity'],pe_fit,"--", color = "red")

# Save the figure
#plt.savefig("../output_data/Fig4.png")

# Show plot
plt.show()
```
